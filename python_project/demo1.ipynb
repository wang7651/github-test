```python
import numpy as np
import matplotlib.pyplot as plt

# 定義正弦波的參數
frequency = 1  # 頻率 (Hz)
amplitude = 1  # 振幅
time = np.linspace(0, 1, 1000)  # 時間軸 (從 0 到 1 秒，共 1000 個點)

# 產生正弦波
sine_wave = amplitude * np.sin(2 * np.pi * frequency * time)

# 繪製圖形
plt.plot(time, sine_wave)
plt.xlabel("時間 (秒)")
plt.ylabel("振幅")
plt.title("正弦波")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定義正弦波的參數
frequency = 1  # 頻率 (Hz)
amplitude = 1  # 振幅
time = np.linspace(0, 1, 1000)  # 時間軸 (從 0 到 1 秒，共 1000 個點)

# 產生正弦波
sine_wave = amplitude * np.sin(2 * np.pi * frequency * time)

# 繪製圖形
plt.plot(time, sine_wave)
plt.xlabel("時間 (秒)")
plt.ylabel("振幅")
plt.title("正弦波")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定義餘弦波的參數
frequency = 1  # 頻率 (Hz)
amplitude = 1  # 振幅
time = np.linspace(0, 1, 1000)  # 時間軸 (從 0 到 1 秒，共 1000 個點)

# 產生餘弦波
cosine_wave = amplitude * np.cos(2 * np.pi * frequency * time)

# 繪製圖形
plt.plot(time, cosine_wave)
plt.xlabel("時間 (秒)")
plt.ylabel("振幅")
plt.title("餘弦波")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成 x 軸數據 (0 到 2π 之間)
x = np.linspace(0, 2 * np.pi, 1000)

# 計算 sin(x) 和 cos(x)
y_sin = np.sin(x)
y_cos = np.cos(x)

# 繪製圖形
plt.figure(figsize=(8, 5))  # 設定圖形大小
plt.plot(x, y_sin, label="sin(x)", color="b", linewidth=2)  # sin(x) 藍色線
plt.plot(x, y_cos, label="cos(x)", color="r", linestyle="--", linewidth=2)  # cos(x) 紅色虛線

# 添加標題與標籤
plt.title("Sin & Cos 波形", fontsize=14)
plt.xlabel("x 軸 (弧度)", fontsize=12)
plt.ylabel("y 軸", fontsize=12)
plt.axhline(0, color="black", linewidth=1)  # x 軸
plt.axvline(0, color="black", linewidth=1)  # y 軸
plt.legend()  # 顯示圖例
plt.grid(True)  # 顯示格線

# 顯示圖形
plt.show()
```
